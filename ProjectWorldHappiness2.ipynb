## 1. Introduction to World Happiness Report

The World Happiness Report is a landmark survey of the state of global happiness. The first report was published in 2012. The World Happiness 2017, which ranks 155 countries by their happiness levels, was released at the United Nations at an event celebrating International Day of Happiness on March 20th. The report continues to gain global recognition as governments, organizations and civil society increasingly use happiness indicators to inform their policy-making decisions. 

Leading experts across fields – economics, psychology, survey analysis, national statistics, health, public policy and more – describe how measurements of well-being can be used effectively to assess the progress of nations. The reports review the state of happiness in the world today and show how the new science of happiness explains personal and national variations in happiness. 

This servey is important, because it help to delve deeper into issues relating to happiness, including mental illness, the objective benefits of happiness, the importance of ethics, policy implications.

World Happiness Report official site https://worldhappiness.report/

## 2. Data description and objectives

To analyse the happiness rank of 155 countries  used dataset from Kaggle(https://www.kaggle.com/unsdsn/world-happiness/notebooks) 2016-2019 years based on seven factors including economy (GDP per capita), social support, health (life expectancy), generosity, trust (government corruption), freedom, and dystopia residual.


Nationally representative samples of respondents are asked to think of a ladder, with the best possible life for them being a 10, and the worst possible life being a 0. They are then asked to rate their own current lives on that 0 to 10 scale. The report correlates the results with various life factors.

Data: 
* Happiness Score - sum of scores of previous parameters
* Happiness Rank - position in the list
* Region - geographical location (only in 2015-2016)
* Economy (GDP per Capita) - measure of a country's economic output that accounts for its number of people
* Social support - having friends and other people, including family, to turn to in times of need or crisis to give you a broader focus and positive self-image 
* Health (Life Expectancy) - measure based on an estimate of the average age  of a particular population group will be when they die.
* Freedom - freedom of making life choices
* Generosity - the quality of being kind and generous
* Trust (Government Corruption) - the Corruption Perceptions Index (CPI), index published annually by Transparency International which ranks countries by their perceived levels of public sector corruption, as determined by expert assessments and opinion surveys

### 4.4. Q4: Analyze historical behaivour of countries 


Columns with year were added to see the historical behaivour and make the comparison easier

```python
d2015['Year'] = 2015 #for all rows it will write year's given value
d2016['Year'] = 2016
d2017['Year'] = 2017 
d2018['Year'] = 2018
d2019['Year'] = 2019 
```

```python
#data preparation and filtering

data1=d2015.filter(['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Year'],axis=1)

data2=d2016.filter(['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Year'],axis=1)

data3=d2017.filter(['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Year'],axis=1)

data4=d2018.filter(['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Year'],axis=1)

data5=d2019.filter(['Country','Happiness Rank','Happiness Score','Economy (GDP per Capita)','Year'],axis=1)

#appending dataframes to data1
data1=data1.append([data2,data3,data4,data5]) 
```

A new dataframe with data of 2015-2019 GDP per capita of every country was created. There were selected several countries from different continents as a sample, such as US(North America), Finland(Europe), Australia, Brasil(South America), China(Asia) and Egypt(Africa)

```python
plt.figure(figsize=(10,8))

df = data1[data1['Country']=='United States']
sns.lineplot(x="Year", y="Economy (GDP per Capita)",data=df,label='US')

df = data1[data1['Country']=='Australia']
sns.lineplot(x="Year", y="Economy (GDP per Capita)",data=df,label='Australia')

df = data1[data1['Country']=='Finland']
sns.lineplot(x="Year", y="Economy (GDP per Capita)",data=df,label='Finland')


df = data1[data1['Country']=='Brazil']
sns.lineplot(x="Year", y="Economy (GDP per Capita)",data=df,label='Brazil')

df = data1[data1['Country']=='China']
sns.lineplot(x="Year", y="Economy (GDP per Capita)",data=df,label='China')

df = data1[data1['Country']=='Egypt']
sns.lineplot(x="Year", y="Economy (GDP per Capita)",data=df,label='Egypt')

plt.title("GDP per capita 2015-2019")
plt.show()
```

It is interesting to note that these countries have similar trends, despite the fact that their values for GDP per capita are different. It could be explained by the world economic factors in certain years. All countries have a peak value in 2017 and a dramatic decrease in 2018, following by a slight increase.
